```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Function to draw 2D diagrams
def draw_2d_design():
    fig, ax = plt.subplots(1, 2, figsize=(12, 6))

    # 2D Front View
    ax[0].add_patch(patches.Rectangle((0, 0), 6, 0.05, edgecolor='blue', facecolor='lightblue', label='Polymer Sheet'))
    ax[0].add_patch(patches.Rectangle((0, 0), 6, 0.05, edgecolor='red', fill=False, linewidth=2, label='Adhesive Area'))
    ax[0].add_patch(patches.Rectangle((6.2, 0), 1.6, 0.05, edgecolor='orange', facecolor='orange', label='Warming Pad'))

    ax[0].set_xlim(-1, 10)
    ax[0].set_ylim(-0.1, 0.2)
    ax[0].set_title("2D Front View")
    ax[0].set_xlabel("Width (cm)")
    ax[0].set_ylabel("Height (cm)")
    ax[0].legend(loc='upper right')
    ax[0].axis('equal')

    # 2D Cross-Section
    ax[1].add_patch(patches.Rectangle((0, 0), 1, 0.005, edgecolor='black', facecolor='cyan', label='Surface Layer'))
    ax[1].add_patch(patches.Rectangle((0, 0.005), 1, 0.03, edgecolor='black', facecolor='lightblue', label='Thermoplastic Polymer'))
    ax[1].add_patch(patches.Rectangle((0, 0.035), 1, 0.005, edgecolor='black', facecolor='gray', label='Adhesive Layer'))

    ax[1].set_xlim(-0.5, 2)
    ax[1].set_ylim(-0.01, 0.05)
    ax[1].set_title("2D Cross-Section View")
    ax[1].set_xlabel("Thickness (cm)")
    ax[1].set_ylabel("Layers")
    ax[1].legend(loc='upper right')
    ax[1].axis('equal')

    plt.tight_layout()
    plt.show()

# Function to draw 3D diagram
def draw_3d_design():
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')

    # 3D Drawing of the Band
    x = np.linspace(-1, 1, 100)
    y = np.linspace(-2, 2, 100)
    X, Y = np.meshgrid(x, y)
    Z = np.zeros_like(X)

    # Draw the thermoplastic polymer sheet
    ax.plot_surface(X, Y, Z, color='lightblue', alpha=0.7)

    # Draw the adhesive areas
    ax.plot_surface(X, Y, Z + 0.0005, color='gray', alpha=0.5)

    # Draw the warming pad
    X2, Y2 = np.meshgrid(np.linspace(1.1, 1.5, 50), y)
    Z2 = np.zeros_like(X2)
    ax.plot_surface(X2, Y2, Z2, color='orange', alpha=0.7)

    # Set labels
    ax.set_title("3D View of the Band")
    ax.set_xlabel("Width")
    ax.set_ylabel("Height")
    ax.set_zlabel("Layers")

    # Set limits
    ax.set_xlim(-2, 2)
    ax.set_ylim(-2, 2)
    ax.set_zlim(0, 0.001)

    plt.show()

# Run the functions to draw the designs
if __name__ == "__main__":
    draw_2d_design()
    draw_3d_design()

    import matplotlib.pyplot as plt
import matplotlib.patches as patches
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Function to draw 2D diagrams
def draw_2d_design():
    fig, ax = plt.subplots(1, 2, figsize=(12, 6))

    # 2D Front View
    ax[0].add_patch(patches.Rectangle((0, 0), 6, 0.05, edgecolor='blue', facecolor='lightblue', label='Polymer Sheet'))
    ax[0].add_patch(patches.Rectangle((0, 0), 6, 0.05, edgecolor='red', fill=False, linewidth=2, label='Adhesive Area'))
    ax[0].add_patch(patches.Rectangle((6.2, 0), 1.6, 0.3, edgecolor='orange', facecolor='orange', label='Warming Pad'))  # 3mm for warming pad

    ax[0].set_xlim(-1, 10)
    ax[0].set_ylim(-0.1, 0.4)
    ax[0].set_title("2D Front View")
    ax[0].set_xlabel("Width (cm)")
    ax[0].set_ylabel("Height (cm)")
    ax[0].legend(loc='upper right')
    ax[0].axis('equal')

    # 2D Cross-Section
    ax[1].add_patch(patches.Rectangle((0, 0), 1, 0.0005, edgecolor='black', facecolor='cyan', label='Surface Layer'))  # 0.5mm
    ax[1].add_patch(patches.Rectangle((0, 0.0005), 1, 0.003, edgecolor='black', facecolor='lightblue', label='Thermoplastic Polymer'))  # 3mm
    ax[1].add_patch(patches.Rectangle((0, 0.0035), 1, 0.0005, edgecolor='black', facecolor='gray', label='Adhesive Layer'))  # 0.5mm

    ax[1].set_xlim(-0.5, 2)
    ax[1].set_ylim(-0.001, 0.004)
    ax[1].set_title("2D Cross-Section View")
    ax[1].set_xlabel("Thickness (cm)")
    ax[1].set_ylabel("Layers")
    ax[1].legend(loc='upper right')
    ax[1].axis('equal')

    # Add grid and labels
    ax[1].yaxis.set_major_locator(plt.MultipleLocator(0.0005))
    ax[1].yaxis.set_minor_locator(plt.MultipleLocator(0.0001))
    ax[1].grid(which='both', linestyle='--', linewidth=0.5)

    plt.tight_layout()
    plt.show()

# Function to draw 3D diagram
def draw_3d_design():
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')

    # 3D Drawing of the Band
    x = np.linspace(-1, 1, 100)
    y = np.linspace(-2, 2, 100)
    X, Y = np.meshgrid(x, y)
    Z = np.zeros_like(X)

    # Draw the thermoplastic polymer sheet
    ax.plot_surface(X, Y, Z, color='lightblue', alpha=0.7)

    # Draw the adhesive areas
    ax.plot_surface(X, Y, Z + 0.0005, color='gray', alpha=0.5)

    # Draw the warming pad
    X2, Y2 = np.meshgrid(np.linspace(1.1, 1.5, 50), y)
    Z2 = np.zeros_like(X2) + 0.003  # 3mm thick warming pad
    ax.plot_surface(X2, Y2, Z2, color='orange', alpha=0.7)

    # Set labels
    ax.set_title("3D View of the Band")
    ax.set_xlabel("Width")
    ax.set_ylabel("Height")
    ax.set_zlabel("Layers")

    # Set limits
    ax.set_xlim(-2, 2)
    ax.set_ylim(-2, 2)
    ax.set_zlim(0, 0.004)

    plt.show()

# Run the functions to draw the designs
if __name__ == "__main__":
    draw_2d_design()
    draw_3d_design()

```
